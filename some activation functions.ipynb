```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(-2.25,2.25,0.1)

def gaussian(x):
  return np.e**(-1*pow(x,2))

y = [gaussian(x) for x in X]
plt.plot(X,y, color="xkcd:lilac")
```

```python
def sinc(x):
  if x == 0:
    out = 1
  out = np.sin(x) / x 
  return out


X = np.arange(-50,50,0.1)

y = [sinc(x) for x in X]
plt.plot(X,y, color="xkcd:spring green")
# print(X)
```

```python
def segmoid(x):
  return 1 / (1 + (np.e**(-x)))

X = np.arange(-6,6,0.1)
y = [segmoid(x) for x in X]
plt.plot(X,y, color="xkcd:terra cotta")
```

```python
X = np.arange(-6,6,0.1)
y = [np.sin(x)  for x in X]
z = [np.cos(x) for x in X]
plt.plot(X,y, color="xkcd:bright green")
plt.plot(X,z, color="xkcd:robin's egg blue")
```

```python
def silu(x):
  return x*segmoid(x)

X = np.arange(-4,2,0.1)

y = [segmoid(x)  for x in X]
z = [silu(x) for x in X]
plt.plot(X,y, color="xkcd:dirty yellow")
plt.plot(X,z, color="xkcd:dusty purple")
```
