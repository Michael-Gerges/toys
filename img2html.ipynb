```python
import sys
from PIL import Image

def imgtohtml(img):
  image = Image.open(img)
  image = image.convert('RGB')
  pixels = image.load()
  width, height = image.size
  f = open("ha.html", "w") 


# Start table)
  f.write ('<html><body>')
  f.write ('<table width="%spx" height="%spx" cellpadding="0" cellspacing="0">' % (width, height))

# Loop through rows
  run_length = 1
  for row in range(height):
      f.write( '<tr>')
    
    # Loop through columns
      for col in range(width):

        # If this is not first row (to make sure we get all columns), the last column, and this is part of a run, increase colspan
          if row != 0 and col < width - 1 and pixels[col, row] == pixels[col + 1, row]:
              run_length = run_length + 1
        
        # If there is no run, or this is the end of a run, output this pixel (with no newline)
          else:
              color = '#%x%x%x' % pixels[col, row]
              colspan = 'colspan="%s"' % run_length if run_length > 1 else ''
              f.write ('<td bgcolor="%s" width="%spx" height="1px" %s></td>' % (color, run_length, colspan))
              run_length = 1

      f.write( '</tr>')

  f.write ('</table>')
  f.write ('</body></html>')
  f.close()
```
